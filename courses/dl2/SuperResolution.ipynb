## Model

```python
# standard conv block (without batchnorm)
# batchnorm changes things; we want as little original pixel change as possible
def conv(ni, nf, kernel_size=3, actn=False):
    layers = [nn.Conv2d(ni, nf, kernel_size, padding=kernel_size//2)]
    if actn: layers.append(nn.ReLU(True))
    return nn.Sequential(*layers)
```

```python
class ResSequential(nn.Module):
    def __init__(self, layers, res_scale=1.0):
        super().__init__()
        self.res_scale = res_scale
        self.m = nn.Sequential(*layers)

    # runs input through a sequential model and adds result to the input
    def forward(self, x): return x + self.m(x) * self.res_scale
    # res_scale -> multiplying by value (.1 or .2) helps stabilize training w/ large batch sizes (weights go to inf)
```

```python
# residual block w/ 2 convs and relu only after the first
def res_block(nf):
    return ResSequential(
        [conv(nf, nf, actn=True), conv(nf, nf)],
        0.1)
```

**transpose convolutions:**  
- add white padding (0 value) around each pixel in the original
- e.g. 3x3 -> 7x7
- do convolution over intermediate to upsample (-> 5x5)

**nearest neighbor upsampling:**  
- same as above except copy pixel value instead of white padding
- better but still not good enough

**pixel shuffle:**  
- RxR conv w/ R*R channels (e.g. 3x3x9)
- shuffle pixels - take pixel from each channel to form 3x3 superpixel in output (upscales by 3)

```python
# not using transpose convolutions/nearest neighbor upsampling -> have checkerboard artifacts!
def upsample(ni, nf, scale):
    layers = []
    # we're upsampling by a scale of 2 -> we need to do that log base 2 scale times...
    # the binary logarithm (log2 n) is the power to which the number 2 must be raised to obtain the value n
    for i in range(int(math.log(scale,2))):
        # doing a scale 2 upsample -> 2*2 = 4
        # nn.PixelShuffle built into pytorch -> creates a superpixel
        layers += [conv(ni, nf*4), nn.PixelShuffle(2)]
    return nn.Sequential(*layers)
```

```python
# SuperResolution Resnet
class SrResnet(nn.Module):
    def __init__(self, nf, scale):
        super().__init__()
        features = [conv(3, 64)]                            # conv 3 -> 64 channels   'richen up the space'
        for i in range(8): features.append(res_block(64))   # 8 res blocks -> 64 channels   'lots of computation'
        features += [conv(64,64), upsample(64, 64, scale),  # upsampling (batchnorm - helpful to scale last layer)
                     nn.BatchNorm2d(64),
                     conv(64, 3)]
        self.features = nn.Sequential(*features)
        
    def forward(self, x): return self.features(x)
```

## Perceptual loss

```python
# pixelShuffle still results in some checkerboard artifacts
# randomly initialize only one RxR kernel and copy it R*R times -> all channels are the same
# initially all RxR superpixels will be the same!
def icnr(x, scale=2, init=nn.init.kaiming_normal_):
    # x - weight tensor to be initialized
    # divide by scale**2 because we're going to copy it
    new_shape = [int(x.shape[0] / (scale ** 2))] + list(x.shape[1:])
    subkernel = torch.zeros(new_shape)
    subkernel = init(subkernel)
    subkernel = subkernel.transpose(0, 1)
    subkernel = subkernel.contiguous().view(subkernel.shape[0],
                                            subkernel.shape[1], -1)
    # make copies of it
    kernel = subkernel.repeat(1, 1, scale ** 2)
    transposed_shape = [x.shape[1]] + [x.shape[0]] + list(x.shape[2:])
    kernel = kernel.contiguous().view(transposed_shape)
    kernel = kernel.transpose(0, 1)
    return kernel
    # returns new weight matrix where each initialized subkernel is repeated scale**2 times
```

```python
m_vgg = vgg16(True)

blocks = [i-1 for i,o in enumerate(children(m_vgg))
              if isinstance(o,nn.MaxPool2d)]
blocks, [m_vgg[i] for i in blocks]
```

```python
vgg_layers = children(m_vgg)[:23]
# just using to compare activations, not training

# m_vgg = nn.Sequential(*vgg_layers).cuda().eval()
m_vgg = nn.Sequential(*vgg_layers).eval()

set_trainable(m_vgg, False)
```

```python
def flatten(x): return x.view(x.size(0), -1)
```

```python
class SaveFeatures():
    features=None
    def __init__(self, m): self.hook = m.register_forward_hook(self.hook_fn)
    def hook_fn(self, module, input, output): self.features = output
    def remove(self): self.hook.remove()        
```

```python
class FeatureLoss(nn.Module):
    def __init__(self, m, layer_ids, layer_wgts):
        super().__init__()
        self.m,self.wgts = m,layer_wgts
        self.sfs = [SaveFeatures(m[i]) for i in layer_ids]

    def forward(self, input, target, sum_layers=True):
        self.m(VV(target.data))  # target -> high res image
        res = [F.l1_loss(input,target)/100]  # pixel loss weighted down (by 100)
        targ_feat = [V(o.features.data.clone()) for o in self.sfs]  # saves perceptual activations from target
        self.m(input)
        res += [F.l1_loss(flatten(inp.features),flatten(targ))*wgt
               for inp,targ,wgt in zip(self.sfs, targ_feat, self.wgts)]
        if sum_layers: res = sum(res)
        return res
    
    def close(self):
        for o in self.sfs: o.remove()
```
